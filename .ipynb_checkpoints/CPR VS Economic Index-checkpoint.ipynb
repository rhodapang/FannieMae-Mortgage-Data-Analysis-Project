```python
National_Monthly_Data = HPI[(HPI['level']=='USA or Census Division')& (HPI['frequency']=='monthly')& (HPI['place_name']=='United States')]
```

```python
National_Monthly_Data = National_Monthly_Data[(National_Monthly_Data['hpi_flavor']== 'purchase-only') & (National_Monthly_Data['hpi_type']=='traditional')]
```

```python
National_Monthly_Data
```

```python
National_Monthly_Data=National_Monthly_Data.rename(columns= {'yr':'Year','period':'Month' }) 
```

```python
National_Monthly_Data.info()
```

```python
National_Monthly_Data['DATE'] = pd.to_datetime(National_Monthly_Data[['Year', 'Month']].assign(DAY=1))
National_Monthly_Data=National_Monthly_Data.drop(columns=['Month','Year'])
```

```python
National_Monthly_Data = National_Monthly_Data.rename(columns={'DATE':'Monthly.Rpt.Prd'})
```

```python
National_Monthly_Data.info()
```

```python
National_Monthly_Data
```

```python
# explore the relationsip between CPR and HPI in United States, so merge the data together to analysis 
df= pd.merge(df, National_Monthly_Data[['Monthly.Rpt.Prd','index_nsa','index_sa']], on ='Monthly.Rpt.Prd',how='left'  )
```

```python
df
```

```python
df = df[7:]
df
```

```python
plt.figure(figsize=(10,8))
plt.plot(df['index_sa'], df['CPR'])
plt.xlabel('hpi_adjusted')
plt.ylabel('CPR')
plt.title('CPR against hpi_adjusted nationawide')
```

```python
plt.figure(figsize=(10,8))
plt.plot(df['index_nsa'], df['CPR'])
plt.title('CPR against hpi_non-adjust')
plt.xlabel('hpi_non-adjusted')
plt.ylabel('CPR')
          
 
```

```python
fig = plt.figure(figsize=(10,8))

ax1 = fig.add_subplot(111)
ax1.plot(df['Monthly.Rpt.Prd'],df['CPR'] , color = 'orange', label = 'CPR')
ax1.set_ylabel('CPR')
plt.legend()

ax2 = ax1.twinx()
ax2.plot(df['Monthly.Rpt.Prd'],df['index_sa'], color = 'blue', label = 'HPI')
ax2.set_ylabel('HPI-adjusted') 
plt.legend()

#plt.legend(handles=[ax1, ax2])
plt.xlabel('monthly_reporting_period')
plt.title('CPR & HPI trend')
plt.show()
```
